```python
import numpy as np

# Variables
y = np.array([70, 65, 90, 95, 110, 115])  # Variable dependiente
x = np.array([80, 100, 120, 140, 160, 180])  # Variable independiente
```

```python
X = np.column_stack((np.ones(len(x)), x))  # Matriz con columna de 1s
Y = y.reshape(-1, 1)  # Vector columna
```

```python
beta = np.linalg.inv(X.T @ X) @ X.T @ Y
intercepto = beta[0][0]
pendiente = beta[1][0]

print(f"Intercepto (β₀): {intercepto}")
print(f"Pendiente (β₁): {pendiente}")
```

```python
import matplotlib.pyplot as plt
```

```python
# Predicciones
y_pred = X @ beta

# Gráfica
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x, y_pred, color='red', label='Línea de regresión')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión lineal simple')
plt.legend()
plt.grid(True)
plt.show()
```

```python
residuos = Y - y_pred
print("Residuos:")
print(residuos.flatten())
```

```python
# SSE: suma de errores al cuadrado
sse = np.sum((Y - y_pred) ** 2)

# SST: total de la varianza de Y
sst = np.sum((Y - np.mean(Y)) ** 2)

# R²
r2 = 1 - (sse / sst)
print(f"R² (coeficiente de determinación): {r2:.4f}")
```

```python
plt.scatter(x, residuos, color='purple')
plt.axhline(y=0, color='black', linestyle='--')
plt.title("Residuos")
plt.xlabel("x")
plt.ylabel("Residuo")
plt.grid(True)
plt.show()
```
